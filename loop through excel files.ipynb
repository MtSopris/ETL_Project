```python
import pandas as pd
from glob import glob, iglob
```

```python
filenames= glob('data/Current_BioDiv/*.xlsx')
filenames
```

```python
all_parks_df=pd.DataFrame()

for f in filenames:
    data = pd.read_excel(f, 'Sheet1')
    all_parks_df = all_parks_df.append(data)
    
all_parks_df.head(620)
    
```

```python
all_parks_df.reset_index(inplace=True)
```

```python
all_parks_df['Park Code'].unique()
```

```python
all_parks_df['Category'].unique()
```

```python
#loops through data and converts it into a dict of lists for each type of catergory found
def myfunction(df):
    anmial=['Mammal', 'Bird', 'Amphibian', 'Fish', 'Vascular Plant', 'Insect',
       'Fungi', 'Non-vascular Plant', 'Other Non-vertebrates',
       'Crab/Lobster/Shrimp', 'Reptile', 'Slug/Snail', 'Spider/Scorpion',
       'Chromista', 'Bacteria']
    parks=df['Park Code'].unique()
    my_dictionary={}
    for each_park in parks: 
        park_dict={}
        park_df=df[df['Park Code']==each_park].reset_index()
        for kind in anmial:
            dict_list =[]
            for i in range(len(park_df)):
                if park_df['Category'][i]== kind: 
                   dict_1 = {
                       'name': park_df['Common Names'][i],
                       'occurance': park_df['Occurrence'][i]
                       'abundance': park_df['Abundance'][i]
                           }
                   dict_list.append(dict_1)
            park_dict[kind] = dict_list
        my_dictionary[each_park]=park_dict
    return my_dictionary
results=myfunction(all_parks_df)
results
```
