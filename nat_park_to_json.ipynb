```python
#loops through data and converts it into a dict of lists for each type of catergory found
def myfunction(df):
    anmial=['Mammal', 'Bird', 'Amphibian', 'Fish', 'Vascular Plant']
    my_dictionary={}
    for kind in anmial:
        dict_list =[]
        for i in range(len(df)):
            if df['Category'][i]== kind:
            
                dict_1 = {
                    'name': df['Common Names'][i],
                    'occurance': df['Occurrence'][i]
                    }
            
                dict_list.append(dict_1)
               
        my_dictionary[kind] = dict_list
    
    return my_dictionary
alag_=myfunction(alag_df)
alag_
```
